### 结果处理

```python
import pandas as pd
import json
eval_path = 'eval_2024-11-28-21-36-20' #'eval_2024-11-23-22-41-20'
data_path = f'./saves/Qwen2-VL-7B-Instruct/lora/{eval_path}/generated_predictions.jsonl'
result_data = []
with open(data_path, 'r', encoding='utf-8') as file:
    for line in file:
        json_data = json.loads(line)
        result_data.append(json_data)
result_df = pd.DataFrame(result_data)
```

```python
result_df.sample(10)
```

```python
predict = []
for text in result_df['predict']:
    text = text.replace('"', '').replace('[', '').replace(']', '').strip()
    if ',' in text:
        text = text.split(',')[0]
        
    predict.append(text)
```

```python
import pandas as pd
# 文件路径
file_path = "data/mire/test1/test1_pic.json"

# 打开文件并加载JSON数据
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

submit_df = pd.DataFrame(data)
```

```python
submit_df['predict'] = predict
```

```python
submit_path = f'./submit/submit_{eval_path}.csv'
submit_df.to_csv(submit_path, encoding="utf-8-sig", index=False)
```
